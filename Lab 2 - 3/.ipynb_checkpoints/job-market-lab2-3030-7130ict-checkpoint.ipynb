```python
#create dataframe for salary ranges
df = pd.read_csv('job-market.csv')
```

```python
#salary ranges plot
fig, (ax1) = plt.subplots(ncols=1, figsize=(10, 15))
df.groupby(['LowestSalary']).count().plot(kind='pie', y='Title', textprops={'fontsize': 14}, ax=ax1)
ax1.set_ylabel('Salary Range', size=22)
plt.show()
```

```python
#OPTIONAL
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('job-market.csv')

locationDf = df[df['Location'].isna()==False]
locationDf = locationDf.groupby('Location').count()
locationDf['Location'] = locationDf.index
locationDf_location = locationDf['Location']
locationDf['Count'] = locationDf["Id"]

locationDf = locationDf['Count']
locationDf_location
```

```python
#optional
%matplotlib inline

from matplotlib import cm

fig, ax = plt.subplots(figsize=(10, 15))
ax.pie(locationDf[0:6], labels = locationDf_location[0:6])

plt.show()
```
